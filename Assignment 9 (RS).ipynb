```python
import warnings
from sklearn.feature_extraction.text import CountVectorizer

# Suppress warnings
warnings.filterwarnings("ignore", category=UserWarning, module="sklearn.feature_extraction.text")

# Convert genres into numerical representation
vectorizer = CountVectorizer(tokenizer=lambda x: x.split(", "))
genre_matrix = vectorizer.fit_transform(file['genre'].fillna(""))
# Check feature names
print(vectorizer.get_feature_names_out())

# Check the resulting matrix
print(genre_matrix.toarray())

```

```python
#Normalize numerical features
from sklearn.preprocessing import MinMaxScaler

# Replace 'Unknown' with NaN, then fill NaN values with 0
file['episodes'] = file['episodes'].replace('Unknown', 0).fillna(0)
file['rating'] = file['rating'].replace('Unknown', 0).fillna(0)

# Convert columns to numeric
file['episodes'] = pd.to_numeric(file['episodes'], errors='coerce').fillna(0)
file['rating'] = pd.to_numeric(file['rating'], errors='coerce').fillna(0)

# Scale the columns
scaler = MinMaxScaler()
file[['episodes', 'rating']] = scaler.fit_transform(file[['episodes', 'rating']])

# Display the first few rows of the updated file
print(file.head())

```

```python
import numpy as np
combined_features=np.hstack([genre_matrix.toarray(),file[['episodes','rating']].values])
combined_features
```

```python
#Recommendation System
from sklearn.metrics.pairwise import cosine_similarity
from scipy.sparse import csr_matrix

# Convert combined_features to a sparse matrix if it's not already sparse
sparse_features = csr_matrix(combined_features)

# Compute cosine similarity using sparse matrices
similarity_matrix = cosine_similarity(sparse_features, dense_output=False)
```
